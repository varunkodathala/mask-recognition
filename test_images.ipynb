```python
import tensorflow as tf
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
model = tf.keras.models.load_model('mask_detection.h5')
```

```python
model.summary()
```

```python
def predict_image(im):
    im = cv2.resize(im,(45,45))
    im = im/255
    im = np.expand_dims(im,axis = 0)
    arr = model.predict(im)
    ids = arr.argmax()
    print(class_labels[ids])
```

```python
class_labels = ['Mask Enabled','Mask Not Availed']
```

```python
im1 = cv2.imread('test3.jpg')
plt.imshow(cv2.cvtColor(im1,cv2.COLOR_BGR2RGB))
predict_image(im1)
```
